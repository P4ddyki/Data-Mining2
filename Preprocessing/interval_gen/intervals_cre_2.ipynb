```python
import pandas as pd
```

```python
orders = pd.read_csv('newOrders_jan.csv', delimiter=',')
```

```python
orders.head(50)
```

```python
interval = list(orders['interval'])
```

```python
ls = list()
```

```python
for t in interval:
    if('-' in t):
        ls.append(-1)
    else:
        ls.append(int(t.split(' ',1)[0]))
```

```python
orders['intervals'] = ls
```

```python
orders.head(30)
```

```python
orders = orders.drop(['comID', 'interval'], axis=1)
```

```python
orders.head(30)
```

```python
orders = orders.drop(columns=orders.columns[0], axis=1)
```

```python
orders.head(30)
```

```python
orders.to_csv('orders_before_jan_new.csv', sep=',', encoding='utf-8')
```
